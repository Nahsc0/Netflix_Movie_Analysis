```python
def summary(df):
    print(f'data shape: {df.shape}')
    summ = pd.DataFrame(df.dtypes, columns=['data type'])
    summ['#missing'] = df.isnull().sum().values 
    summ['%missing'] = df.isnull().sum().values / len(df) * 100
    summ['#unique'] = df.nunique().values
    desc = pd.DataFrame(df.describe(include='all').transpose())
    summ['min'] = desc['min'].values
    summ['max'] = desc['max'].values
    summ['average'] = desc['mean'].values
    summ['standard_deviation'] = desc['std'].values
    summ['first value'] = df.loc[0].values
    summ['second value'] = df.loc[1].values
    summ['third value'] = df.loc[2].values
    
    return summ
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('imdb_movies_shows.csv')

# Task 1: Change genres and country features to be comma-separated values.
df['genres'] = df['genres'].apply(lambda x: ', '.join(eval(x)) if pd.notnull(x) else 'NO_GENRE')
df['production_countries'] = df['production_countries'].apply(lambda x: ', '.join(eval(x)) if pd.notnull(x) else 'NO_COUNTRY')

# Task 2: Format countries values to be characters.
df['production_countries'] = df['production_countries'].apply(lambda x: ''.join([i for i in x if i.isalpha()]))

# Task 3: Change the [] values in genres and production_countries to "NO_GENRE" and "NO_COUNTRY".
df['genres'] = df['genres'].apply(lambda x: 'NO_GENRE' if x == '' else x)
df['production_countries'] = df['production_countries'].apply(lambda x: 'NO_COUNTRY' if x == '' else x)

# Task 4: Check the movie without the title.
print("Movies without title:")
print(df[df['title'].isnull()])

# Task 5: Drop movies without imdb_id.
df = df.dropna(subset=['imdb_id'])

# Task 6: Group movies by type.
adult_ratings = ['TV-MA', 'R', 'NC-17']
supervision_ratings = ['PG', 'TV-14', 'G', 'PG-13', 'TV-PG']

# Convert 'age_certification' to numeric
df['age_certification'] = pd.to_numeric(df['age_certification'], errors='coerce')

# Create 'rating_group' based on age certification
df['rating_group'] = pd.cut(df['age_certification'], bins=[-1, 17, 100], labels=['Requires Supervision', 'Adult'])

# Task 7: Map ratings for children.
children_ratings = ['TV-Y', 'TV-G', 'TV-Y7']
df['rating_group'] = df['age_certification'].map(lambda x: 'Good For Children' if x in children_ratings else x)

# Task 8: Fill NaN values with 'Not Rated'.
df['rating_group'] = df['rating_group'].fillna('Not Rated')

# Display the modified DataFrame
print(df)
```
